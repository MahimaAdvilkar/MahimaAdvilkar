```python
#Making a Data Frame and Reading the CSV file

df= pd.read_csv('/Users/mahimaadvilkar/Desktop/Spotify_Youtube.csv')
```

```python
#DATA Preparation

#Dropping
#Renaming
#Null Values
#Dropping Duplicates

```

```python
#Renaming columns
df= df.rename(columns={'Album_type':'Type_of_Album', 
                       'Title':'Youtube_Title',
                      'Channel': 'Youtube_Channel',
                      'Views':'Youtube_Views',
                      'Likes': 'Youtube_Likes',
                      'Comments': 'Youtube_Comments',
                      'official_video':'Official_Video',
                      'Stream':'Youtube_Stream'})
```

```python
# add new column that contains duration of song in minutes

df['Duration_In_Minutes'] = df['Duration_ms'] / 60000
```

```python
#Checking for Null Values

df.isna().sum()
```

```python
#Dropping Columns 
df.drop(['Unnamed: 0', 'Uri', 'Description'], axis=1, inplace=True)
```

```python
# Identify columns where the number of null values is <= 5% of the DataFrame's length
columns_to_consider = df.columns[df.isna().sum() <= len(df) * 0.05]

# Drop rows with null values in these columns
df.dropna(subset=columns_to_consider, inplace=True)
```

```python
# Convert categorical variables if needed
df['Licensed'] = df['Licensed'].map({True: 'Licensed', False: 'Not Licensed'})
```

```python
#Checking for Null Values

df.isna().sum()
```

```python
# converting data types

cols_to_convert = ['Youtube_Views', 'Youtube_Likes', 'Youtube_Comments', 'Youtube_Stream']

for col in cols_to_convert:
    df[col] = df[col].astype('int')

df['Official_Video'] = df['Official_Video'].astype('str')
df['Official_Video'] = df['Official_Video'].str.replace('True', 'Yes')
df['Official_Video'] = df['Official_Video'].str.replace('False', 'No')
```

```python
#4. EDA

# Creating a figure to hold the visualizations
fig, axes = plt.subplots(5, 2, figsize=(20, 25))  # 5 rows, 2 columns

# Assigning vibrant colors
colors = ['navy', 'darkorange', 'darkgreen', 'crimson', 'purple', 
          'chocolate', 'olive', 'darkred', 'midnightblue', 'indigo']

# Plotting distribution for each attribute with vibrant colors
sns.histplot(df['Danceability'], bins=30, kde=True, ax=axes[0, 0], color=colors[0])
axes[0, 0].set_title('Distribution of Danceability')

sns.histplot(df['Energy'], bins=30, kde=True, ax=axes[0, 1], color=colors[1])
axes[0, 1].set_title('Distribution of Energy')

sns.histplot(df['Key'], bins=30, kde=False, ax=axes[1, 0], color=colors[2])
axes[1, 0].set_title('Distribution of Key')

sns.histplot(df['Loudness'], bins=30, kde=True, ax=axes[1, 1], color=colors[3])
axes[1, 1].set_title('Distribution of Loudness')

sns.histplot(df['Speechiness'], bins=30, kde=True, ax=axes[2, 0], color=colors[4])
axes[2, 0].set_title('Distribution of Speechiness')

sns.histplot(df['Acousticness'], bins=30, kde=True, ax=axes[2, 1], color=colors[5])
axes[2, 1].set_title('Distribution of Acousticness')

sns.histplot(df['Instrumentalness'], bins=30, kde=True, ax=axes[3, 0], color=colors[6])
axes[3, 0].set_title('Distribution of Instrumentalness')

sns.histplot(df['Liveness'], bins=30, kde=True, ax=axes[3, 1], color=colors[7])
axes[3, 1].set_title('Distribution of Liveness')

sns.histplot(df['Valence'], bins=30, kde=True, ax=axes[4, 0], color=colors[8])
axes[4, 0].set_title('Distribution of Valence')

sns.histplot(df['Tempo'], bins=30, kde=True, ax=axes[4, 1], color=colors[9])
axes[4, 1].set_title('Distribution of Tempo')

# Adjusting layout for better presentation
plt.tight_layout()
plt.show()
```

```python
features = df[['Danceability', 'Energy', 'Valence']]
features_standardized = StandardScaler().fit_transform(features)
kmeans = KMeans(n_clusters=3, random_state=42).fit(features_standardized)
df['Cluster'] = kmeans.labels_
```

```python
# Selecting the features for clustering
features = df[['Danceability', 'Energy', 'Valence', 'Speechiness', 'Acousticness']]

# Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Performing K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_scaled)

# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Energy', y='Danceability', hue='Cluster', palette='viridis', alpha=0.6)
plt.title('Clusters based on Energy and Danceability')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.legend(title='Cluster')
plt.show()

# Analyzing Clusters
for cluster in sorted(df['Cluster'].unique()):
    cluster_data = df[df['Cluster'] == cluster]
    print(f"Cluster {cluster} Mean Values:")
    print(cluster_data[['Danceability', 'Energy', 'Valence', 'Speechiness', 'Acousticness']].mean())
    print("-" * 50)
```

```python
# Assuming 'df' is your DataFrame and it contains a column 'Cluster' with the cluster labels
cluster_3_tracks = df[df['Cluster'] == 2]

# Display the tracks in Cluster 3
print(cluster_3_tracks[['Artist', 'Track', 'Album']])
```

```python
# Count the number of tracks in each cluster
cluster_counts = df['Cluster'].value_counts()

# Display the count of tracks in each cluster
print(cluster_counts)

plt.figure(figsize=(8, 6))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette="Set2")
plt.title('Number of Tracks in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count of Tracks')
plt.show()
```

```python
#3. Top Artist by their Engagement Rate


# Calculating engagement rate = (Likes + Comments) / Views * 100
df['Engagement_Rate'] = (df['Youtube_Likes'] + df['Youtube_Comments']) / df['Youtube_Views'] * 100

# Group by 'Artist' and calculate mean engagement rate for each artist
artist_engagement = df.groupby('Artist')['Engagement_Rate'].mean().reset_index()

# Sort by 'Engagement_Rate' in descending order to find top artists
top_artists_by_engagement = artist_engagement.sort_values(by='Engagement_Rate', ascending=False).head(10)

# Visualize the top artists by engagement rate
plt.figure(figsize=(12, 8))
sns.barplot(x='Engagement_Rate', y='Artist', data=top_artists_by_engagement, palette='viridis')
plt.title('Top 10 Artists by YouTube Engagement Rate')
plt.xlabel('Average Engagement Rate (%)')
plt.ylabel('Artist')
plt.show()
```
